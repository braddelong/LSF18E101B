### <font color="000088">Let's Do the Math...</font>

Let's do the math and unpack what is going on here. And let's move into the flexprice model, because the U.S. economy is near full employment, so we are not talking about a cyclical recovery boosting short-run growth by putting lots of unemployed people to work:

* $ \frac{K}{Y} = 4 $
* $ α = 0.4 $
* $Y = K^α(EL)^{1-α} $

To Your iClickers: What is dY/dK?

1. 25%
2. 40%
3. 10%
4. 16%
5. None of the above

&nbsp;

⟹ dY/dK = α(Y/K) = 10%

* Boost growth by 0.34%/year
* GDP of 20 trillion dollars
* What is ΔY?

1. 340 billion
2. 680 billion
3. 170 billion
4. 34 billion
5. None of the above

&nbsp;

⟹ ΔY = 20 trillion x 0.34% = 70 billion

What is ΔI?

1. 700 billion
2. 200 billion
3. 340 billion
4. 900 billion
5. None of the above

&nbsp;

⟹ ΔI = 70 billion/10% = 700 billion

* I/Y = 16%
    * ⟹ ΔI/Y = 3.5%
    * ⟹ ΔI/I = 22%

### <font color="000088">Why Are We Not Surprised Not to See It?</font>

* As I said before, flexprice model...

* Investment needs to be looked at from both the supply and the demand side:
    * $ I = I_o - I_rr = (T - G) + (IM - GX) + (1 - [c_o + c_y(1-t)])Y $ 
    * $ I = I_o - I_rr = (tY - G) + (im_yY - x_fY_f - - x_εε_o - x_εε_rr^f + x_εε_rr) + (1 - [c_o + c_y(1-t)])Y $
    * $ ΔI = ΔI_o - I_rΔr = (1-c_y)YΔt + x_εε_rΔr - YΔc_o $
        * With a tax cut, $ (1-c_y)YΔt $ goes the wrong way
        * $ YΔc_o $ simply does not happen

And so we are left, on the supply side, with:

> $ΔI = x_εΔε = x_εε_rΔr  $

What's going on here? The higher interest rates will induce people to stop buying U.S. exports. People who then earn dollars by selling us imports will have no choice but to invest that money in the United States—and that increased foreign saving will drive higher investment.

&nbsp;

> $ ΔI = x_εε_rΔr x_ε = 500 ε_r = 10 $

By how much would r have to go up to make $ ΔI = 700 $?

### <font color="000088">How Do They Justify Writing This?</font>

* Robert Barro (2017): crickets
* Kevin Hassett (2003): “The U.S. is an open economy the borrows in an international capital market, hence... little or no impact on interest rates…”
* Greg Mankiw (2017): “An open economy has the production function y = f(k), where y is output per worker and k is capital per worker. The capital stock adjusts so that the after-tax marginal product of capital equals the exogenously given world interest rate r…”

What does this mean? This means: $ r = r^f $

* Why does $ r = r^f$? We know that:
    * $ (r = r^f) ⟹ (ε = ε_o) $
    * $ Δr = 0 and Δε = 0 $
    * $ ΔI = x_εΔε = x_εε_rΔr $

What must be going on for both of these things to be true? What are they assuming behind the curtain?

1. $ x_ε = 0 $
2. $ ε_rΔr $
3. $ x_ε = ∞ $
4. None of the above

⟹ xεΔ = ∞

* Often heard (especially from Hassett): “The U.S. is an open economy the borrows in an international capital market, hence [policy change X] will have little of no impact on interest rates…”
    * This misses the fact that international capital mobility requires international imbalances in trade…
    * And that requires movements in exchange rates
    * Which then support differences in interest rates across countries…

### <font color="000088">Effects of Capital Inflows</font>

Reference: **Olivier Blanchard, Jonathan D. Ostry, Atish R. Ghosh, and Marcos Chamon** (2015): _[Expansionary or contractionary effects of capital inflows: It depends what kind](https://voxeu.org/article/macro-effects-capital-inflows-capital-type-matters)_: "Some scholars view capital inflows as contractionary, but many policymakers view them as expansionary. Evidence supports the policymakers.... For a given policy rate, bond inflows lead to currency appreciation and are contractionary, while non-bond inflows lead to an appreciation but also to a decrease in the cost of borrowing, and thus may be expansionary..."

----

A _bond inflow_ is: a $ {\Delta}{\epsilon}_o < 0 $

A _non-bond inflow_ (DFI) is:

* a $ {\Delta}{\epsilon}_o < 0 $
* a $ {\Delta}{\rho} < 0 $
    * where $ r = i + \rho - \pi $

So holding i constant:

> $ {\Delta}Y = {\mu}(x_{\epsilon}{\Delta}{\epsilon}_o - (I_r + x_{\epsilon}{\epsilon}_r){\Delta}{\rho}) $

The key question is: when capital flows in, does it just come by itself or does it bring an increased supply of risk-bearing capacity along with it?

----

Reference: Basic Model:

> $ Y = C + I + G + NX $

> $ C = c_o + c_y(1-t)Y $

> $ I = I_o - I_rr $

> $ G $

> $ NX = GX - IM $

> $ IM = im_y $

> $ GX = x_fY^f + x_{\epsilon}{\epsilon} $

> $ {\epsilon} = {\epsilon}_o + {\epsilon}_r(r^f - r) $

> $ r = i + {\rho} - {\pi} $

&nbsp;

> $ MPE = c_y(1-t) - im_y $

> $ \mu = \frac{1}{1 - MPE} $

> $ A_o = [c_o + I_o + G] + [x_fY^f + x_{\epsilon}{\epsilon}_o + x_{\epsilon}{\epsilon}_rr^f] $

> $ Y = \mu(A_o - (I_r + x_{\epsilon}{\epsilon}_r)r) $

&nbsp;

### <font color="000088">Effects of Capital Outflows</font>

The reverse of the effects of capital inflows:

A _bond outflow_ is: a $ {\Delta}{\epsilon}_o > 0 $

A _non-bond outflow_ is:

* a $ {\Delta}{\epsilon}_o > 0 $
* a $ {\Delta}{\rho} > 0 $
    * where $ r = i + \rho - \pi $

So holding i constant:

> $ {\Delta}Y = {\mu}(x_{\epsilon}{\Delta}{\epsilon}_o - (I_r + x_{\epsilon}{\epsilon}_r){\Delta}{\rho}) $

The key question is: when capital flows out, does it just leave by itself or does it carry risk-bearing capacity away with it?
